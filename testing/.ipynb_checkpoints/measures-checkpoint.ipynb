## Measures Evaluated

### AAB — Avoidance of Antibiotics for Acute Bronchitis/Bronchiolitis

**What it is:** An overuse measure that tracks whether members diagnosed with acute bronchitis were prescribed antibiotics — which are not recommended for this viral condition.

**What it measures:**
- Denominator: Members aged 18+ with an acute bronchitis/bronchiolitis episode (ICD-10: J20, J21) who did not have a comorbidity that might justify antibiotics
- Numerator: Members who were NOT prescribed antibiotics within 3 days before or after the diagnosis
- Direction: Higher is better (high rate = good — fewer unnecessary antibiotics)

**Why it matters:** Antibiotic overuse drives antimicrobial resistance, a major public health crisis. Acute bronchitis is viral in the vast majority of cases; antibiotics provide no benefit and cause harm.

**Data required:** ICD-10 codes (J20, J21) from carrier and outpatient; PDE or carrier pharmacy claims to identify antibiotic prescriptions.

```python
pd.read_sql("""
    WITH eligible AS (
        SELECT BENE_ID
        FROM beneficiary
        WHERE BENE_DEATH_DT IS NULL
          AND FLOOR(DATEDIFF(day, CONVERT(date, BENE_BIRTH_DT, 106), '2021-12-31') / 365.25) >= 18
    )
    SELECT
        (SELECT COUNT(*) FROM eligible) AS eligible_population,
        COUNT(DISTINCT d.BENE_ID) AS members_with_bronchitis_dx
    FROM (
        SELECT BENE_ID, PRNCPAL_DGNS_CD AS dx FROM outpatient
            WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
        UNION ALL
        SELECT BENE_ID, LINE_ICD_DGNS_CD FROM carrier
            WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
    ) d
    JOIN eligible e ON d.BENE_ID = e.BENE_ID
    WHERE d.dx LIKE 'J20%' OR d.dx LIKE 'J21%'
""", conn)
```

**Conclusion:** KEEP
7,817 eligible members (18+). 795 members with acute bronchitis diagnosis in 2021.
Strong denominator. Proceeding with implementation.

### ABA — Adult BMI Assessment

**What it is:** A preventive care measure that tracks whether adult members had their BMI documented and received a follow-up plan if their BMI was outside the normal range.

**What it measures:**
- Denominator: Members aged 18–74 with at least one outpatient visit during the measurement year
- Numerator: Members who had BMI documented AND received a follow-up plan if BMI was < 18.5 or >= 25
- Direction: Higher is better

**Why it matters:** BMI documentation at preventive and primary care visits is a basic quality indicator. It is a gateway to obesity counseling and intervention, which reduces risk of diabetes, cardiovascular disease, and other chronic conditions.

**Data required:** HCPCS codes from carrier and outpatient to identify qualifying office visits. Key codes: 99385-99387 (preventive), 99395-99397 (preventive established), 99201-99215 (E&M visits).

```python
pd.read_sql("""
    WITH eligible AS (
        SELECT BENE_ID
        FROM beneficiary
        WHERE BENE_DEATH_DT IS NULL
          AND FLOOR(DATEDIFF(day, CONVERT(date, BENE_BIRTH_DT, 106), '2021-12-31') / 365.25)
              BETWEEN 18 AND 74
    )
    SELECT
        (SELECT COUNT(*) FROM eligible) AS eligible_population,
        COUNT(DISTINCT h.BENE_ID) AS members_with_office_visit_codes
    FROM (
        SELECT BENE_ID, HCPCS_CD FROM carrier
            WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
        UNION ALL
        SELECT BENE_ID, HCPCS_CD FROM outpatient
            WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
    ) h
    JOIN eligible e ON h.BENE_ID = e.BENE_ID
    WHERE h.HCPCS_CD IN (
        '99385','99386','99387','99395','99396','99397',
        '99201','99202','99203','99204','99205',
        '99211','99212','99213','99214','99215'
    )
""", conn)
```

**Conclusion:** KEEP
5,958 eligible members (ages 18-74). 341 members with qualifying office visit codes in 2021.
Strong denominator. Proceeding with implementation.

### AMR — Asthma Medication Ratio

**What it is:** A medication management measure that tracks whether members with persistent asthma are on an appropriate ratio of controller to total asthma medications.

**What it measures:**
- Denominator: Members aged 5–64 with persistent asthma (ICD-10: J45) and at least one asthma medication dispensed
- Numerator: Members with a ratio of controller medications to total asthma medications of 0.50 or greater
- Direction: Higher is better — a higher ratio indicates appropriate use of controllers over relievers

**Why it matters:** Overreliance on short-acting bronchodilators (relievers) at the expense of inhaled corticosteroids (controllers) is associated with poor asthma control, hospitalizations, and mortality.

**Data required:** ICD-10 codes (J45) from claims to identify persistent asthma; PDE pharmacy data to identify and classify controller vs reliever medications by NDC code.

```python
pd.read_sql("""
    WITH eligible AS (
        SELECT BENE_ID
        FROM beneficiary
        WHERE BENE_DEATH_DT IS NULL
          AND FLOOR(DATEDIFF(day, CONVERT(date, BENE_BIRTH_DT, 106), '2021-12-31') / 365.25)
              BETWEEN 5 AND 64
    ),
    asthma_dx AS (
        SELECT DISTINCT BENE_ID
        FROM (
            SELECT BENE_ID, PRNCPAL_DGNS_CD AS dx FROM inpatient
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
            UNION ALL
            SELECT BENE_ID, PRNCPAL_DGNS_CD FROM outpatient
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
            UNION ALL
            SELECT BENE_ID, LINE_ICD_DGNS_CD FROM carrier
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
        ) d
        WHERE dx LIKE 'J45%'
    )
    SELECT
        (SELECT COUNT(*) FROM eligible) AS eligible_population,
        (SELECT COUNT(DISTINCT a.BENE_ID) FROM asthma_dx a JOIN eligible e ON a.BENE_ID = e.BENE_ID)
            AS members_with_asthma_dx,
        (
            SELECT COUNT(DISTINCT a.BENE_ID)
            FROM asthma_dx a
            JOIN eligible e ON a.BENE_ID = e.BENE_ID
            JOIN pde p ON a.BENE_ID = p.BENE_ID
            WHERE CONVERT(date, p.SRVC_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
        ) AS asthma_members_with_any_pde_2021
""", conn)
```

**Conclusion:** KEEP
3,674 eligible members (ages 5-64). 220 members with asthma diagnosis in 2021.
157 of those members have PDE records in 2021, enabling the controller/reliever ratio calculation.
Proceeding with implementation.

### BCS — Breast Cancer Screening

**What it is:** A cancer screening measure that tracks whether eligible female members received a mammogram during the measurement year or the prior year.

**What it measures:**
- Denominator: Female members aged 52–74 as of Dec 31, 2021
- Numerator: Members with a mammogram (2D or 3D) in the measurement year or year prior
- Direction: Higher is better

**Why it matters:** Breast cancer is the most common cancer in women. Regular mammography is the primary evidence-based screening tool for early detection.

**Data required:** HCPCS codes from carrier and outpatient. Key codes: 77067 (screening mammography, bilateral), 77066, 77065.

```python
pd.read_sql("""
    WITH eligible AS (
        SELECT BENE_ID
        FROM beneficiary
        WHERE BENE_DEATH_DT IS NULL
          AND SEX_IDENT_CD = '2'
          AND FLOOR(DATEDIFF(day, CONVERT(date, BENE_BIRTH_DT, 106), '2021-12-31') / 365.25)
              BETWEEN 52 AND 74
    )
    SELECT
        (SELECT COUNT(*) FROM eligible) AS eligible_population,
        COUNT(DISTINCT h.BENE_ID) AS members_with_mammography_codes
    FROM (
        SELECT BENE_ID, HCPCS_CD FROM carrier
            WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2020-01-01' AND '2021-12-31'
        UNION ALL
        SELECT BENE_ID, HCPCS_CD FROM outpatient
            WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2020-01-01' AND '2021-12-31'
    ) h
    JOIN eligible e ON h.BENE_ID = e.BENE_ID
    WHERE h.HCPCS_CD IN ('77067','77066','77065')
""", conn)
```

**Conclusion:** DROP
2,071 eligible members (female, 52-74). Only 2 members with mammography codes across 2020-2021.
Mammography HCPCS codes are nearly absent from the synthetic dataset. Numerator would be effectively zero.
Not implementable.

### CDC — Comprehensive Diabetes Care, HbA1c Testing

**What it is:** A chronic disease management measure that tracks whether members with diabetes received an HbA1c test during the measurement year.

**What it measures:**
- Denominator: Members aged 18–75 with a diabetes diagnosis (ICD-10: E10, E11)
- Numerator: Members who received an HbA1c test (CPT: 83036, 83037)
- Direction: Higher is better

**Why it matters:** HbA1c testing is the standard measure of long-term blood glucose control for diabetes management. Untested members may have poorly controlled diabetes without detection.

**Data required:** ICD-10 codes (E10, E11) for denominator; lab CPT codes (83036, 83037) for numerator. Lab claims are required — not guaranteed in all claims datasets.

```python
pd.read_sql("""
    WITH eligible AS (
        SELECT BENE_ID
        FROM beneficiary
        WHERE BENE_DEATH_DT IS NULL
          AND FLOOR(DATEDIFF(day, CONVERT(date, BENE_BIRTH_DT, 106), '2021-12-31') / 365.25)
              BETWEEN 18 AND 75
    ),
    diabetes_dx AS (
        SELECT DISTINCT BENE_ID
        FROM (
            SELECT BENE_ID, PRNCPAL_DGNS_CD AS dx FROM inpatient
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
            UNION ALL
            SELECT BENE_ID, PRNCPAL_DGNS_CD FROM outpatient
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
            UNION ALL
            SELECT BENE_ID, LINE_ICD_DGNS_CD FROM carrier
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
        ) d
        WHERE dx LIKE 'E10%' OR dx LIKE 'E11%'
    )
    SELECT
        (SELECT COUNT(*) FROM eligible) AS age_eligible_population,
        (SELECT COUNT(DISTINCT h.BENE_ID) FROM diabetes_dx h JOIN eligible e ON h.BENE_ID = e.BENE_ID)
            AS members_with_diabetes_dx,
        (
            SELECT COUNT(DISTINCT BENE_ID)
            FROM (
                SELECT BENE_ID, HCPCS_CD FROM carrier
                    WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
                UNION ALL
                SELECT BENE_ID, HCPCS_CD FROM outpatient
                    WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
            ) h
            WHERE HCPCS_CD IN ('83036','83037')
        ) AS members_with_hba1c_codes
""", conn)
```

**Conclusion:** DROP
6,138 age-eligible members. 423 members with a diabetes diagnosis in 2021.
However, HbA1c test codes (83036, 83037) return 0 members across carrier and outpatient for all years,
not just 2021. Lab claims are not included in the CMS synthetic FFS dataset.
The numerator cannot be constructed. Not implementable.

### COL — Colorectal Cancer Screening

**What it is:** A cancer screening measure that tracks whether eligible members received a recommended colorectal cancer screening during the measurement year or in the look-back period.

**What it measures:**
- Denominator: Members aged 45–75 as of Dec 31, 2021
- Numerator: Members with a qualifying colorectal screening — colonoscopy, flexible sigmoidoscopy, CT colonography, FIT-DNA test, or FOBT
- Direction: Higher is better

**Why it matters:** Colorectal cancer is the second leading cause of cancer death in the U.S. Regular screening catches polyps and early-stage cancer before symptoms appear.

**Data required:** HCPCS codes from carrier and outpatient claims. Key codes: 45378 (colonoscopy), 45380, 45385, G0328 (FOBT), 82274 (FIT), 82270.

```python
pd.read_sql("""
    WITH eligible AS (
        SELECT BENE_ID
        FROM beneficiary
        WHERE BENE_DEATH_DT IS NULL
          AND FLOOR(DATEDIFF(day, CONVERT(date, BENE_BIRTH_DT, 106), '2021-12-31') / 365.25)
              BETWEEN 45 AND 75
    )
    SELECT
        (SELECT COUNT(*) FROM eligible) AS eligible_population,
        COUNT(DISTINCT h.BENE_ID) AS members_with_col_codes
    FROM (
        SELECT BENE_ID, HCPCS_CD FROM carrier
            WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
        UNION ALL
        SELECT BENE_ID, HCPCS_CD FROM outpatient
            WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
    ) h
    JOIN eligible e ON h.BENE_ID = e.BENE_ID
    WHERE h.HCPCS_CD IN ('45378','45380','45385','G0328','82274','82270')
""", conn)
```

**Conclusion:** KEEP
4,860 eligible members. 724 members with colonoscopy or FOBT codes in 2021.
Strong denominator and numerator coverage. Proceeding with implementation.

### FUH — Follow-Up After Hospitalization for Mental Illness

**What it is:** A follow-up measure that tracks whether members discharged from an inpatient psychiatric hospitalization receive outpatient mental health follow-up within 7 and 30 days.

**What it measures:**
- Denominator: Members aged 6+ discharged from an inpatient stay with a principal mental illness diagnosis (ICD-10: F20-F99)
- Numerator (7-day): Members with an outpatient mental health visit within 7 days of discharge
- Numerator (30-day): Members with an outpatient mental health visit within 30 days of discharge
- Direction: Higher is better

**Why it matters:** Psychiatric readmissions are most likely in the first 30 days post-discharge. Timely follow-up is the single most effective intervention to prevent readmission after a mental health hospitalization.

**Data required:** Inpatient claims with mental illness diagnosis codes to identify the hospitalization; carrier and outpatient claims to identify follow-up visits.

```python
pd.read_sql("""
    WITH mh_hospitalizations AS (
        -- Inpatient discharges in 2021 with a principal mental illness diagnosis
        SELECT DISTINCT BENE_ID, CONVERT(date, CLM_THRU_DT, 106) AS discharge_dt
        FROM inpatient
        WHERE CLM_LINE_NUM = 1
          AND CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
          AND (
              PRNCPAL_DGNS_CD LIKE 'F2%' OR PRNCPAL_DGNS_CD LIKE 'F3%'
              OR PRNCPAL_DGNS_CD LIKE 'F4%' OR PRNCPAL_DGNS_CD LIKE 'F5%'
              OR PRNCPAL_DGNS_CD LIKE 'F6%' OR PRNCPAL_DGNS_CD LIKE 'F7%'
              OR PRNCPAL_DGNS_CD LIKE 'F8%' OR PRNCPAL_DGNS_CD LIKE 'F9%'
          )
    )
    SELECT
        (SELECT COUNT(*) FROM mh_hospitalizations) AS mh_discharges_2021,
        (
            SELECT COUNT(DISTINCT h.BENE_ID)
            FROM mh_hospitalizations h
            JOIN (
                SELECT BENE_ID, CONVERT(date, CLM_FROM_DT, 106) AS svc_dt FROM carrier
                UNION ALL
                SELECT BENE_ID, CONVERT(date, CLM_FROM_DT, 106) FROM outpatient
            ) f ON h.BENE_ID = f.BENE_ID
            WHERE f.svc_dt > h.discharge_dt
              AND f.svc_dt <= DATEADD(day, 7, h.discharge_dt)
        ) AS followup_within_7_days
""", conn)
```

**Conclusion:** KEEP
19 members with a mental illness inpatient discharge in 2021. 9 with follow-up within 7 days.
Follow-up claims are detectable in carrier and outpatient tables.
Both the 7-day and 30-day numerator windows are implementable.
Proceeding with implementation.

### FUM — Follow-Up After Emergency Department Visit for Mental Illness

**What it is:** A follow-up measure that tracks whether members who visit the emergency department for a mental health condition receive outpatient follow-up care within 7 days.

**What it measures:**
- Denominator: Members with an ED visit with a principal mental illness diagnosis (ICD-10: F20-F99) who were not admitted inpatient from the ED
- Numerator: Members who had an outpatient mental health visit within 7 days of the ED discharge
- Direction: Higher is better

**Why it matters:** ED visits for mental illness often represent a crisis. Without timely follow-up, patients are at high risk of deterioration and repeat visits. This measure is a natural companion to FUH.

**Data required:** Outpatient claims with ED revenue center codes (0450-0459, 0981) and mental illness diagnosis codes; carrier and outpatient claims to identify follow-up visits.

```python
pd.read_sql("""
    WITH mental_health_outpatient AS (
        SELECT DISTINCT BENE_ID
        FROM (
            SELECT BENE_ID, PRNCPAL_DGNS_CD AS dx FROM outpatient
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
            UNION ALL
            SELECT BENE_ID, ICD_DGNS_CD1 FROM outpatient
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
        ) d
        WHERE dx LIKE 'F2%' OR dx LIKE 'F3%' OR dx LIKE 'F4%' OR dx LIKE 'F5%'
           OR dx LIKE 'F6%' OR dx LIKE 'F7%' OR dx LIKE 'F8%' OR dx LIKE 'F9%'
    ),
    mental_health_carrier AS (
        SELECT DISTINCT BENE_ID
        FROM carrier
        WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
          AND (
              LINE_ICD_DGNS_CD LIKE 'F2%' OR LINE_ICD_DGNS_CD LIKE 'F3%'
              OR LINE_ICD_DGNS_CD LIKE 'F4%' OR LINE_ICD_DGNS_CD LIKE 'F5%'
              OR LINE_ICD_DGNS_CD LIKE 'F6%' OR LINE_ICD_DGNS_CD LIKE 'F7%'
              OR LINE_ICD_DGNS_CD LIKE 'F8%' OR LINE_ICD_DGNS_CD LIKE 'F9%'
          )
    )
    SELECT
        (SELECT COUNT(*) FROM mental_health_outpatient) AS members_with_mh_outpatient_2021,
        (SELECT COUNT(*) FROM mental_health_carrier)    AS members_with_mh_carrier_2021
""", conn)
```

**Conclusion:** KEEP
205 members with mental health outpatient claims in 2021. 102 with mental health carrier claims.
Both the ED visit population and the follow-up visit population are detectable in the data.
Proceeding with implementation.

### IET — Initiation and Engagement of Substance Use Disorder Treatment

**What it is:** A two-part measure that tracks whether members newly diagnosed with a substance use disorder begin treatment and then engage in ongoing treatment.

**What it measures:**
- Denominator: Members aged 13+ with a new SUD diagnosis (ICD-10: F10-F19) with no SUD treatment in the prior 60 days
- Numerator Part 1 (Initiation): Members who receive SUD treatment within 14 days of the index diagnosis
- Numerator Part 2 (Engagement): Members who receive two or more additional SUD treatment services within 34 days of initiation
- Direction: Higher is better for both parts

**Why it matters:** SUD is one of the most undertreated conditions in healthcare. The gap between diagnosis and treatment initiation is a major quality problem. IET measures whether the system is connecting people with treatment at the critical moment of diagnosis.

**Data required:** ICD-10 codes (F10-F19) from any claim type to identify new SUD episodes; HCPCS codes from carrier and outpatient to identify SUD treatment services.

```python
pd.read_sql("""
    WITH sud_diagnosis AS (
        SELECT BENE_ID, MIN(CONVERT(date, CLM_FROM_DT, 106)) AS first_sud_dt
        FROM (
            SELECT BENE_ID, CLM_FROM_DT, PRNCPAL_DGNS_CD AS dx FROM outpatient
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
            UNION ALL
            SELECT BENE_ID, CLM_FROM_DT, ICD_DGNS_CD1 FROM outpatient
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
            UNION ALL
            SELECT BENE_ID, CLM_FROM_DT, LINE_ICD_DGNS_CD FROM carrier
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
            UNION ALL
            SELECT BENE_ID, CLM_FROM_DT, PRNCPAL_DGNS_CD FROM inpatient
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
        ) d
        WHERE dx LIKE 'F1%'
        GROUP BY BENE_ID
    ),
    followup AS (
        SELECT DISTINCT s.BENE_ID
        FROM sud_diagnosis s
        JOIN (
            SELECT BENE_ID, CONVERT(date, CLM_FROM_DT, 106) AS svc_dt FROM outpatient
            UNION ALL
            SELECT BENE_ID, CONVERT(date, CLM_FROM_DT, 106) FROM carrier
        ) f ON s.BENE_ID = f.BENE_ID
        WHERE f.svc_dt > s.first_sud_dt
          AND f.svc_dt <= DATEADD(day, 34, s.first_sud_dt)
    )
    SELECT
        (SELECT COUNT(*) FROM sud_diagnosis) AS members_with_new_sud_dx_2021,
        (SELECT COUNT(*) FROM followup)      AS sud_members_with_followup_within_34_days
""", conn)
```

**Conclusion:** KEEP
370 members with a new SUD diagnosis (F10-F19) in 2021. 238 with any follow-up claim within 34 days.
Strong denominator and detectable follow-up activity. Both initiation and engagement windows are implementable.
Proceeding with implementation.

### LBP — Use of Imaging Studies for Low Back Pain

**What it is:** An overuse measure that tracks whether members with acute low back pain received unnecessary imaging (X-ray, MRI, or CT scan) within 28 days of diagnosis.

**What it measures:**
- Denominator: Members aged 18–50 with a new acute low back pain episode (ICD-10: M54.x) without a prior episode in the past 6 months
- Numerator: Members who received lumbar imaging within 28 days
- Direction: Lower is better — routine imaging for uncomplicated acute LBP is not clinically indicated

**Why it matters:** Inappropriate imaging for low back pain is one of the most well-documented forms of low-value care. It drives cost without improving outcomes and exposes patients to unnecessary radiation.

**Data required:** ICD-10 codes (M54.x) to identify the episode; lumbar imaging HCPCS codes from carrier and outpatient (72100 series for X-ray, 72141 series for MRI, 72125 series for CT).

```python
pd.read_sql("""
    WITH aged AS (
        SELECT BENE_ID,
               FLOOR(DATEDIFF(day, CONVERT(date, BENE_BIRTH_DT, 106), '2021-12-31') / 365.25) AS age
        FROM beneficiary
        WHERE BENE_DEATH_DT IS NULL
    ),
    lbp_members AS (
        SELECT DISTINCT d.BENE_ID
        FROM (
            SELECT BENE_ID, PRNCPAL_DGNS_CD AS dx FROM outpatient
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
            UNION ALL
            SELECT BENE_ID, ICD_DGNS_CD1 FROM outpatient
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
            UNION ALL
            SELECT BENE_ID, LINE_ICD_DGNS_CD FROM carrier
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
        ) d
        JOIN aged a ON d.BENE_ID = a.BENE_ID
        WHERE d.dx LIKE 'M54%' AND a.age BETWEEN 18 AND 50
    )
    SELECT
        (SELECT COUNT(*) FROM lbp_members) AS members_with_lbp_dx,
        (
            SELECT COUNT(DISTINCT BENE_ID)
            FROM (
                SELECT BENE_ID, HCPCS_CD FROM carrier
                    WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
                UNION ALL
                SELECT BENE_ID, HCPCS_CD FROM outpatient
                    WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
            ) h
            WHERE h.BENE_ID IN (SELECT BENE_ID FROM lbp_members)
              AND HCPCS_CD IN (
                  '72100','72102','72110','72114','72120',
                  '72141','72142','72146','72147','72148','72156','72157','72158',
                  '72125','72126','72127','72128','72129','72130','72131','72132'
              )
        ) AS lbp_members_with_imaging
""", conn)
```

**Conclusion:** DROP
124 members with low back pain diagnosis (ages 18-50). 0 members with lumbar imaging codes.
Lumbar spine imaging HCPCS codes are absent from the synthetic dataset — same pattern as BCS.
Not implementable.

### PCR — Plan All-Cause Readmissions

**What it is:** A utilization measure that tracks how often members are readmitted to the hospital within 30 days of a prior inpatient discharge.

**What it measures:**
- Denominator: All inpatient discharges in the measurement year (index admissions), excluding planned readmissions and transfers
- Numerator: Index admissions followed by an unplanned inpatient readmission within 30 days
- Direction: Lower is better — a high readmission rate signals poor care transitions or premature discharge

**Why it matters:** Readmission reduction has been a major focus of CMS payment policy since the Hospital Readmissions Reduction Program (HRRP) in 2012. It reflects the quality of care coordination at discharge and in the community.

**Data required:** Inpatient claims only. Uses CLM_FROM_DT (admission date) and CLM_THRU_DT (discharge date) to identify index admissions and calculate the 30-day window.

```python
pd.read_sql("""
    WITH index_admissions AS (
        -- One row per inpatient claim in 2021 (CLM_LINE_NUM = 1 avoids counting each line separately)
        SELECT
            BENE_ID,
            CONVERT(date, CLM_FROM_DT, 106) AS admit_dt,
            CONVERT(date, CLM_THRU_DT, 106) AS discharge_dt
        FROM inpatient
        WHERE CLM_LINE_NUM = 1
          AND CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
    ),
    readmissions AS (
        -- For each index admission, look for any inpatient admission within 30 days of discharge
        SELECT DISTINCT i.BENE_ID, i.admit_dt
        FROM index_admissions i
        JOIN inpatient r
            ON i.BENE_ID = r.BENE_ID
            AND r.CLM_LINE_NUM = 1
            AND CONVERT(date, r.CLM_FROM_DT, 106) > i.discharge_dt
            AND CONVERT(date, r.CLM_FROM_DT, 106) <= DATEADD(day, 30, i.discharge_dt)
    )
    SELECT
        COUNT(*) AS index_admissions_2021,
        (SELECT COUNT(*) FROM readmissions) AS readmissions_within_30_days,
        ROUND(
            CAST((SELECT COUNT(*) FROM readmissions) AS FLOAT) / COUNT(*) * 100, 1
        ) AS readmission_rate_pct
    FROM index_admissions
""", conn)
```

**Conclusion:** KEEP
3,049 index admissions in 2021. 955 readmissions within 30 days (31.3% rate).
The readmission rate is higher than the real-world average (~15%), which is a known characteristic
of synthetic data — the Synthea generator does not model realistic care transition patterns.
The data is fully present and the measure is implementable. Proceeding with implementation.

### PPC — Prenatal and Postpartum Care

**What it is:** A prenatal care measure that tracks whether female members who delivered a live birth received a prenatal visit in the first trimester and a postpartum visit within 7–84 days after delivery.

**What it measures:**
- Denominator: Female members aged 8–64 who had a live birth delivery during the measurement year
- Numerator: Members with a prenatal visit in the first trimester (or within 42 days of enrollment) and a postpartum visit within 7–84 days after delivery
- Direction: Higher is better

**Why it matters:** Prenatal care is one of the most effective preventive interventions in medicine. Early prenatal visits reduce rates of low birth weight, preterm birth, and maternal complications.

**Data required:** Delivery CPT codes from carrier/outpatient to identify the index delivery event. Key codes: 59400 (vaginal delivery), 59510 (C-section), 59610, 59618.

```python
pd.read_sql("""
    WITH eligible AS (
        SELECT BENE_ID
        FROM beneficiary
        WHERE BENE_DEATH_DT IS NULL
          AND SEX_IDENT_CD = '2'
          AND FLOOR(DATEDIFF(day, CONVERT(date, BENE_BIRTH_DT, 106), '2021-12-31') / 365.25)
              BETWEEN 8 AND 64
    )
    SELECT
        (SELECT COUNT(*) FROM eligible) AS eligible_population,
        COUNT(DISTINCT h.BENE_ID) AS members_with_delivery_codes
    FROM (
        SELECT BENE_ID, HCPCS_CD FROM carrier
            WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
        UNION ALL
        SELECT BENE_ID, HCPCS_CD FROM outpatient
            WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
    ) h
    JOIN eligible e ON h.BENE_ID = e.BENE_ID
    WHERE h.HCPCS_CD IN ('59400','59510','59610','59618')
""", conn)
```

**Conclusion:** DROP
1,737 eligible members (female, 8-64). 0 members with delivery codes in 2021.
Delivery CPT codes are completely absent from the synthetic dataset.
Not implementable.

### SAA — Adherence to Antipsychotic Medications for Individuals with Schizophrenia

**What it is:** A pharmacy adherence measure that tracks whether members with schizophrenia stay on antipsychotic medications throughout the measurement year.

**What it measures:**
- Denominator: Members aged 18–64 with schizophrenia (ICD-10: F20-F29) who had at least two encounters and at least one antipsychotic prescription during the year
- Numerator: Members with a Proportion of Days Covered (PDC) >= 80% for antipsychotic medications
- Direction: Higher is better — consistent adherence is critical for managing schizophrenia and preventing relapse

**Why it matters:** Schizophrenia is a severe, chronic condition where medication non-adherence is the leading cause of relapse and rehospitalization. PDC >= 80% is the standard adherence threshold across pharmacy-based HEDIS measures.

**Data required:** ICD-10 codes (F20-F29) from claims to identify the population; PDE pharmacy records to calculate PDC.

```python
pd.read_sql("""
    WITH aged AS (
        SELECT BENE_ID,
               FLOOR(DATEDIFF(day, CONVERT(date, BENE_BIRTH_DT, 106), '2021-12-31') / 365.25) AS age
        FROM beneficiary
        WHERE BENE_DEATH_DT IS NULL
    ),
    schizophrenia_members AS (
        SELECT DISTINCT d.BENE_ID
        FROM (
            SELECT BENE_ID, PRNCPAL_DGNS_CD AS dx FROM inpatient
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
            UNION ALL
            SELECT BENE_ID, PRNCPAL_DGNS_CD FROM outpatient
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
            UNION ALL
            SELECT BENE_ID, LINE_ICD_DGNS_CD FROM carrier
                WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
        ) d
        JOIN aged a ON d.BENE_ID = a.BENE_ID
        WHERE d.dx LIKE 'F2%' AND a.age BETWEEN 18 AND 64
    )
    SELECT
        (SELECT COUNT(*) FROM schizophrenia_members) AS members_with_schizophrenia_dx,
        (
            SELECT COUNT(DISTINCT s.BENE_ID)
            FROM schizophrenia_members s
            JOIN pde p ON s.BENE_ID = p.BENE_ID
            WHERE CONVERT(date, p.SRVC_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
        ) AS schizophrenia_members_with_pde_2021
""", conn)
```

**Conclusion:** DROP
0 members with schizophrenia-spectrum diagnosis (F20-F29) in 2021.
Schizophrenia codes are not present in the synthetic dataset, despite the broader mental illness
category (F20-F99) being well represented for FUH and FUM.
Not implementable.

### URI — Upper Respiratory Infection

**What it is:** An overuse measure that tracks whether members diagnosed with a nonspecific upper respiratory infection were inappropriately prescribed antibiotics.

**What it measures:**
- Denominator: Members aged 3 months+ with an upper respiratory infection episode (ICD-10: J00, J06) without a comorbidity that justifies antibiotics
- Numerator: Members who were NOT prescribed antibiotics within 3 days
- Direction: Higher is better

**Why it matters:** Like AAB, this measure targets antibiotic overuse. URI is almost always viral. Antibiotic prescribing for URI is a primary driver of antimicrobial resistance.

**Data required:** ICD-10 codes (J00, J06) from claims; PDE or carrier pharmacy claims to identify antibiotic prescriptions.

```python
pd.read_sql("""
    SELECT COUNT(DISTINCT BENE_ID) AS members_with_uri_dx_2021
    FROM (
        SELECT BENE_ID, PRNCPAL_DGNS_CD AS dx FROM outpatient
            WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
        UNION ALL
        SELECT BENE_ID, ICD_DGNS_CD1 FROM outpatient
            WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
        UNION ALL
        SELECT BENE_ID, LINE_ICD_DGNS_CD FROM carrier
            WHERE CONVERT(date, CLM_FROM_DT, 106) BETWEEN '2021-01-01' AND '2021-12-31'
    ) d
    WHERE dx LIKE 'J00%' OR dx LIKE 'J06%'
""", conn)
```

**Conclusion:** DROP
Only 40 members with a URI diagnosis in 2021 across all claim types.
Population is too thin to produce a meaningful rate. Not implementable.
